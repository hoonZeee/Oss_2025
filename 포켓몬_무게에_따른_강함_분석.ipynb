**포켓몬은 무게가 많이 나갈수록 강할까?**

```
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# CSV 파일을 pandas DataFrame으로 읽어옵니다.
df = pd.read_csv('/content/pokedex.csv')

# 몸무게 기준으로 데이터를 분석합니다.
print("몸무게 통계:")
print(df['weight'].describe())

# 몸무게가 가장 큰 포켓몬 출력
print("\n몸무게가 가장 큰 포켓몬:")
heaviest_pokemon = df.loc[df['weight'].idxmax()]
print(heaviest_pokemon[['name', 'weight']])

# 몸무게가 가장 작은 포켓몬 출력
print("\n몸무게가 가장 작은 포켓몬:")
lightest_pokemon = df.loc[df['weight'].idxmin()]
print(lightest_pokemon[['name', 'weight']])

# 몸무게 기준으로 포켓몬 정렬하여 출력
print("\n몸무게 순으로 정렬된 포켓몬:")
sorted_by_weight = df.sort_values('weight', ascending=False)
print(sorted_by_weight[['name', 'weight']])

# 추가적인 분석 (예: 몸무게 분포 히스토그램)
# df['weight_kg'].hist()
# plt.xlabel('몸무게 (kg)')
# plt.ylabel('포켓몬 수')
# plt.title('포켓몬 몸무게 분포')
# plt.show()
```

```
# 몸무게와 공격력, 방어력의 시각화

import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot for weight vs attack
plt.figure(figsize=(10, 6))
sns.scatterplot(x='weight', y='attack', data=df)
plt.xlabel('Weight')
plt.ylabel('Attack')
plt.title('Relationship between Weight and Attack')
plt.show()

# Histogram for weight distribution
plt.figure(figsize=(10, 6))
plt.hist(df['weight'], bins=20)
plt.xlabel('Weight')
plt.ylabel('Number of Pokémon')
plt.title('Distribution of Pokémon Weights')
plt.show()

# Scatter plot for weight vs defense
plt.figure(figsize=(10, 6))
sns.scatterplot(x='weight', y='defense', data=df)
plt.xlabel('Weight')
plt.ylabel('Defense')
plt.title('Relationship between Weight and Defense')
plt.show()
```

```
# Visualize the relationship between weight, attack, and defense using scatter plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x='weight', y='attack', data=df)
plt.xlabel('Weight')
plt.ylabel('Attack')
plt.title('Relationship between Weight and Attack')

plt.subplot(1, 2, 2)
sns.scatterplot(x='weight', y='defense', data=df)
plt.xlabel('Weight')
plt.ylabel('Defense')
plt.title('Relationship between Weight and Defense')

plt.tight_layout()
plt.show()

# Analyze the correlation between weight, attack, and defense
correlation = df[['weight', 'attack', 'defense']].corr()
print("Correlation between Weight, Attack, and Defense:")
print(correlation)

# Scatter plot with regression line for weight vs attack
sns.lmplot(x='weight', y='attack', data=df)
plt.xlabel('Weight')
plt.ylabel('Attack')
plt.title('Relationship between Weight and Attack (Regression Line)')
plt.show()

# Scatter plot with regression line for weight vs defense
sns.lmplot(x='weight', y='defense', data=df)
plt.xlabel('Weight')
plt.ylabel('Defense')
plt.title('Relationship between Weight and Defense (Regression Line)')
plt.show()


```
